Label Encoding

```python
from sklearn import preprocessing
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
labelencoder = LabelEncoder()
df["smoking_status"]=labelencoder.fit_transform(df["smoking_status"])
df["ever_married"]=labelencoder.fit_transform(df["ever_married"])
df["work_type"]=labelencoder.fit_transform(df["work_type"])
df["Residence_type"]=labelencoder.fit_transform(df["Residence_type"])
df["gender"]=labelencoder.fit_transform(df["gender"])
```

```python
df.head()
```

```python
#Correlation
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(),cmap='Blues',annot=True)
```

OneHot Encoding

```python
oe =OneHotEncoder(sparse=False,handle_unknown="error",drop="first")
oe
```

```python
ohe = OneHotEncoder()
X = ohe.fit_transform(df.work_type.values.reshape(-1,1)).toarray()
OHE = pd.DataFrame(X, columns = ["Private","Self-employed","Govt_job","children","Never_worked"])
df = pd.concat([df, OHE], axis=1)
df = df.drop("work_type",axis = 1)
```

```python
X = ohe.fit_transform(df.smoking_status.values.reshape(-1,1)).toarray()
OHE = pd.DataFrame(X, columns = ["smokes","Unknown","formerly smoked","never smoked"])
df = pd.concat([df, OHE], axis=1)
df = df.drop("smoking_status",axis = 1)
```

```python
X = ohe.fit_transform(df.ever_married.values.reshape(-1,1)).toarray()
OHE = pd.DataFrame(X, columns = ["No","Yes"])
df = pd.concat([df, OHE], axis=1)
df = df.drop("ever_married",axis = 1)
```

```python
X = ohe.fit_transform(df.Residence_type.values.reshape(-1,1)).toarray()
OHE = pd.DataFrame(X, columns = ["Rural","Urban"])
df = pd.concat([df, OHE], axis=1)
df = df.drop("Residence_type",axis = 1)
```

```python
X = ohe.fit_transform(df.gender.values.reshape(-1,1)).toarray()
OHE = pd.DataFrame(X, columns = ["Male","Female","Other"])
df = pd.concat([df, OHE], axis=1)
df = df.drop("gender",axis = 1)
```

Dividing Training and Testing parts of the data

```python
X = df.drop('stroke', axis=1)
y = df['stroke']
```

```python
X.head()
```

```python
y.head()
```

### Feature scaling

Scaling the features is to avoid intensive computation and also avoid one variable dominating the others. For a binary classification problem, no need to scale the dependent variable. But for regression, we need to scale the dependent variables.

```python
from sklearn.preprocessing import StandardScaler
scaler= StandardScaler()
X=scaler.fit_transform(X)
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, train_size=0.8, shuffle=True, random_state=0)
```

```python
X_train.shape, y_train.shape
```

```python
X_test.shape, y_test.shape
```

### Model Building

Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import plot_roc_curve
lr = LogisticRegression()
lr.fit(X_train, y_train)
```

```python
#Make prediction
predict=lr.predict(X_test)
predict
```

```python
y_test
```

```python
#Traing Score 
lr.score(X_train,y_train)
```

```python
#Testing score
lr.score(X_test,y_test)*100
```

```python
print('Accuracy score:',accuracy_score(y_test,predict))
```

Evaluation for logisitic regression

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(confusion_matrix(y_test,predict))
```

```python
print(classification_report(y_test,predict))
```

```python
plot_roc_curve(lr,X_test,y_test)
```

Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
DTC= DecisionTreeClassifier(max_depth=3)
DTC.fit(X_train,y_train)
```

```python
y_pre=DTC.predict(X_test)
y_pre
```

```python
#Training Score
DTC.score(X_train,y_train)
```

```python
#Testing Score
DTC.score(X_test,y_test)*100
```

```python
print('Accuracy score',accuracy_score(y_test,y_pre))
```

```python
print(confusion_matrix(y_test,y_pre))
print(classification_report(y_test,y_pre))
```

```python
plot_roc_curve(DTC,X_test,y_test)
```

Support Vector Classifier

```python
plot_roc_curve(svc,X_test,y_test)
```
